```python
df1 = pd.DataFrame({'count' : df.groupby( [ "Country", "classification",'Name'] ).size()}).reset_index()
```

```python
df11 =df.groupby(['Country','classification']).agg({ 'Name':'count'}).reset_index()
```

```python
df11
```

```python
# Imports
from scipy.stats import stats, ttest_1samp, ttest_rel, ttest_ind

# Null hypotheses
# good cities for nomads have more unicorns

#split the data
good_city = df11[(df11['classification']=='good')]['Name']
bad_city = df11[(df11['classification']=='bad')]['Name']

#confidence
alpha = 0.05

#test
result = ttest_ind(good_city,bad_city)

pval = result.pvalue

if pval <0.05:
  print('Reject the null hypothesis as ', pval ,' is smaller than aplha ', alpha)
else:
  print("We can't reject the null hypothesis as ", pval, ' is higher than aplha ', alpha)
```
